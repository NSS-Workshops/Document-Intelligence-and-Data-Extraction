# Using spaCy for Unstructured Documents

In this notebook, you'll work with the spaCy library to parse information from unstructured documents.

First, we need to load the spaCy library and a language model.

```python
import spacy
nlp = spacy.load("en_core_web_sm")

# Wrap output in Colab
from IPython.display import HTML, display
def set_css():
  display(HTML('''
  <style>
    pre {
      white-space: pre-wrap;
    }
  </style>
  '''))

get_ipython().events.register('pre_run_cell', set_css)
```

We'll work with the following paragraph.

```python
text = "On March 15, 2024, Apple Inc. announced that it would invest $2 billion in a new artificial intelligence research center in Austin, Texas. CEO Tim Cook said the facility is expected to create over 3,000 jobs in the next five years. The project will focus on machine learning, data privacy, and advanced hardware development, partnering with researchers from the University of Texas."
print(text)
```

## Step 1: Convert Text to a SpaCy Document

```python
doc = nlp(text)
```

## Step 2: Inspect Sentences

spaCy has divided the text we provided into sentences. We can acces these through the sents attribute.


```python
for sent in doc.sents:
  print(sent)
  print('-----')
```

## Step 3: Inspect Tokens

spaCy also divides the text into tokens, which can be words, numbers, or punctuation.

```python
for token in doc:
    print(token.text, end=" | ")
```

## Step 4: Look at Part-of-Speech Tags

```python
print("\n\nToken : Part-of-Speech")
for token in doc:
    print(f"{token.text:15} : {token.pos_}")
```

## Step 5: Inspect Named Entities and Labels

```python
print("\nNamed Entities:")
for ent in doc.ents:
    print(f"{ent.text:30} : {ent.label_}")
```

## Step 6: View the Dependency Using Displacy

```python
from spacy import displacy

displacy.render(doc, style="dep", jupyter=True, options={"distance": 100})
```

#### Director

Here's a starting function to extract the director. Try it out on different movies and see where it works and where it fails. Then attempt to improve the logic so that it is more accurate.

```python
def extract_director(text):
    doc = nlp(text)

    directors = []

    for sent in doc.sents:
        if "directed by" in sent.text.lower():
            for ent in sent.ents:
                if ent.label_ == "PERSON":
                    directors.append(ent.text)

    return directors
```

#### Title

Again, here is a starting function to attempt to extract titles. Read through it to understand the logic, and then check to see how well it works.

```python
def extract_title(text):
    # Parse the input text with spaCy
    doc = nlp(text)

    # Get the first sentence of the text
    # We assume the movie title appears in the subject of the first sentence
    first_sentence = list(doc.sents)[0]

    # Loop through all tokens (words) in the first sentence
    for token in first_sentence:
        # Check if this token is the grammatical subject ("nsubj") of the sentence
        # The subject often corresponds to the movie title in a sentence like
        # "Alien is a 1979 science fiction film directed by Ridley Scott."
        if token.dep_ == "nsubj":
            # Get all tokens that are part of this subject's subtree
            # The subtree includes the token itself and all words attached to it
            # For example, in "The Lord of the Rings: The Return of the King",
            # the subtree captures the full title, not just the first word
            subtree_tokens = list(token.subtree)

            # Find the start and end positions of the subtree within the document
            start = subtree_tokens[0].i         # index of the first token in the subtree
            end = subtree_tokens[-1].i + 1      # index after the last token (Python slicing convention)

            # Create a span in the doc covering all the tokens in the subtree
            title_span = doc[start:end]

            # Return the text of this span as the movie title
            return title_span.text

    # If no subject is found, return None
    return None
```

```python
i = 0
movie = list(movies.keys())[i]
print(f'Movie Title: {movie}')
print('----------')
print(f'Extracted Title: {extract_title(movies[movie])}')
print('-----------')
print(movies[movie])
```

#### Release Date

```python
# Regex patterns for dates
date_patterns = [
    r'\b(?:January|February|March|April|May|June|July|August|September|October|November|December)\s\d{1,2},\s\d{4}\b',
    r'\b\d{1,2}\s(?:January|February|March|April|May|June|July|August|September|October|November|December)\s\d{4}\b'
]

def extract_release_date(text):
    doc = nlp(text)

    for sent in doc.sents:
        # Look only at sentences mentioning release/premiere/opened
        if any(word in sent.text.lower() for word in ["release", "premiere", "opened"]):
            for pattern in date_patterns:
                match = re.search(pattern, sent.text)
                if match:
                    return match.group(0)

    return None
```

```python
i = 0
movie = list(movies.keys())[i]
print(f'Movie Title: {movie}')
print('----------')
print(f'Extracted Release Date: {extract_release_date(movies[movie])}')
print('-----------')
print(movies[movie])
```
